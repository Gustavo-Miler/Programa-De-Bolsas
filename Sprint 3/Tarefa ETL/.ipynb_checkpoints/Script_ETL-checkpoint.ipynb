```python
#EXTRAÇÃO

with open('actors.csv') as arquivo:
    linhas = arquivo.readlines()
    col_actor = []
    col_total_gross = []
    col_num_movies = []
    col_avg_movie = []
    col_best_movie = []
    col_gross = []
    
    for linha in linhas[1:]:
        if '"' in linha:
            nome = linha.split('"')[1]
            atual = linha.split('"')[2].split(',')
            atual[0] = nome
        else:
            atual = linha.split(',')
        col_actor.append(atual[0])
        col_total_gross.append(atual[1])
        col_num_movies.append(int(atual[2]))
        col_avg_movie.append(float(atual[3].split(' ')[0]))
        col_best_movie.append(atual[4])
        col_gross.append(float(atual[5]))
    
    '''#TAREFA 1
    #ATOR/ATRIZ COM MAIOR NUMERO DE FILMES E RESPECTIVO NUMERO DE FILMES
    maior = max(col_num_movies)
    index = col_num_movies.index(maior)
    ator = col_actor[index]
    
    output = f'Ator: {ator} número de filmes: {maior}'
    
    with open('Tarefa_1.txt', 'w') as f:
        f.write(output)'''
    
    '''#TAREFA 2
    #MEDIA DE NUMERO DE FILMES POR ATOR
    media = sum(col_num_movies) / len(col_num_movies)
    output = f'A média de filmes por ator é de {media}'
    
    with open('Tarefa_2.txt', 'w') as f:
        f.write(output)'''
    
    '''#TAREFA 3
    #ATOR COM MAIOR MÉDIA POR FILME
    maior = max(col_avg_movie)
    index = col_avg_movie.index(maior)
    ator = col_actor[index]
    output = f'O ator/atriz com a maior média por filme é {ator} com {maior} por filme'
    
    with open('Tarefa_3.txt', 'w') as f:
        f.write(output)'''
    
    '''#TAREFA 4
    #NOME DOS FILMES MAIS FREQUENTES E SUA RESPECTIVA FREQUENCIA
    frequente = max(set(col_best_movie), key=col_best_movie.count)
    frequencia = col_best_movie.count(frequente) / len(col_best_movie)
    output = f'O filme mais frequente e sua respectiva frequência é {frequente} com {frequencia} de frequencia'
    
    with open('Tarefa_4.txt', 'w') as f:
        f.write(output)'''
    
    #TAREFA 5
    #LISTA DOS AUTORES ORDENADA DO MAIS BEM PAGO AO MENOS BEM PAGO
    li = []
    lista = col_total_gross
    for i in range(len(lista)):
        li.append([lista[i], i])
    li.sort(reverse=True)
    sort_index = []
    
    for x in li:
        sort_index.append(x[1])
    
    output = ''
    for i in range(len(lista)):
        output += f'Ator: {col_actor[sort_index[i]]} Pagamento: {lista[sort_index[i]]}\n'
    
    with open('Tarefa_5.txt', 'w') as f:
        f.write(output)
```
